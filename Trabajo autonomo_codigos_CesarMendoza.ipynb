```python
def dividir(a, b):
    return a / b
print(dividir(7,3))
```

```python
def promedio(lista):
    return sum(lista)/len(lista)
print(promedio([-1, 2, 3]))
```

```python
def obtener_elemento(lista, posicion):
    return lista[posicion]
print(obtener_elemento([10, 20, 30], -1))
```

```python
def es_par(numero):
    return numero % 2 == 0
print(es_par(0))
```

```python
def convertir_a_entero(cadena):
    return int(cadena)
print(convertir_a_entero("4.5"))
```

```python
def concatenar(a, b):
    return a + b
print(concatenar(1,2)) 
```

```python
def buscar_valor(diccionario, clave):
    return diccionario[clave]
print(buscar_valor({1: "uno"}, 1))
```

```python
def validar_positivo(numero): 
    if numero <= 0: raise ValueError("Debe ser positivo") 
    return numero
print(validar_positivo("abc"))
```

```python
def invertir_cadena(cadena): 
    return cadena[::-1] 
print(invertir_cadena("hola mundo"))
```

```python
def factorial(n): if n == 0: 
    return 1 
    return n * factorial(n-1)
print(factorial(2.5))
```
